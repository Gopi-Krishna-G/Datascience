```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1 = np.linspace(start = -3, stop = 3, num = 300)
```

```python
def f(x):
    return x**2 + x + 1
```

```python
plt.figure(figsize = (10,8))
plt.title('SIMPLE COST FUNCTION', fontsize = 20)
plt.xlabel('X', fontsize = 18)
plt.ylabel('f(x)', fontsize = 18)
plt.grid()
plt.plot(x_1, f(x_1))
plt.show()
```

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize = (10,8))
plt.title('DERIVATIVE OF THE SIMPLE COST FUNCTION', fontsize = 20)
plt.xlabel('X', fontsize = 18)
plt.ylabel('df(x)', fontsize = 18)
plt.grid()
plt.plot(x_1, df(x_1))
plt.show()
```
